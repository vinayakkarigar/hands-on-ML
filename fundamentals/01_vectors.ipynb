# Length of a vector

```python
# Load the data. Here we are taking Iris dataset as our example
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np

iris_ds = tfds.load('iris', split='train')
# get a sample of size N from the ds
N = 6
iris_ex = iris_ds.take(N)
tf.print(iris_ex)
# Array to store the dataset samples
tf_arr = tf.TensorArray(tf.float32,  size = 0, dynamic_size=True, clear_after_read=False)
tf_arr
```

```python
# Calculate length of the vectors
index = 0
for sample in iris_ex:
  feat = sample["features"]
  label = sample["label"]
  # sample to array
  tf_arr = tf_arr.write(index, feat)
  index += 1
  vector_norm = tf.norm(feat)      #alculate the lenght of vector. Treat "norm" and "length" as synonym.
  tf.print("Vector: ", feat, "Vector Norm: ", vector_norm)
```

# Distance between 2 vectors
Remember whenever we say vectors imagine the space of physical vectors, polynomial, functions, matrices, converging sequences, electric fields etc. Thus you should be able to define the distance between 2 elements/vectors belonging to all these spaces.

Let u and v are the 2 vectors then distance is a metric defined as below
distance( u , v ) = || u - v ||   here (u -v) is a vector and ||.|| is the notation for norm
hence the distance is the norm of the difference vector
